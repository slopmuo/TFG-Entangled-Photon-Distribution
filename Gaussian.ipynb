```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Creamos solo un gráfico (1 fila, 1 columna) y ajustamos el tamaño
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(-3.5, 3.5, 400)

# Gaussian N(0, 1°)
ax.plot(x, norm.pdf(x, 0, 1), 'k-', linewidth=2)
ax.fill_between(x, norm.pdf(x, 0, 1), where=(np.abs(x) <= 1),
                alpha=0.3, color='steelblue', label=r'$|\delta| \leq 1°$ (~68%)')
ax.fill_between(x, norm.pdf(x, 0, 1), where=(np.abs(x) <= 2),
                alpha=0.15, color='steelblue', label=r'$|\delta| \leq 2°$ (~95%)')

ax.set_xlabel(r'$\delta$ (degrees)', fontsize=11)
ax.set_ylabel('Probability density', fontsize=11)
ax.set_title(r'Gaussian $\mathcal{N}(0°, 1°)$', fontsize=12)
ax.set_xlim(-3.5, 3.5)
ax.axvline(0, color='gray', lw=0.8, ls='--')
ax.legend(fontsize=9)

plt.tight_layout()
# Se cambió el nombre del archivo guardado para que tenga sentido
plt.savefig('gaussian.pdf', dpi=180, bbox_inches='tight')
plt.show()
```
